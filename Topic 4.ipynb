```r
#PRINCIPAL COMPONENT ANALISYS (PCA)
my_cars <- data.frame(mtcars)
my_cars
summary(my_cars)
cor(my_cars)
```

```r
# matrix diagram
x <- as.matrix(my_cars)
pairs(x, gap=0, diag.panel=function(x) {
  par(new=TRUE)
  hist(x, col="light pink", probability=TRUE)
  lines(density(x), col="red", lwd=1)
})
```

```r
#PCA by princomp()
pca.my_cars <- princomp(my_cars[, 1:7], scores=TRUE)
pca.my_cars
summary(pca.my_cars)
loadings(pca.my_cars)

plot(pca.my_cars)
```

```r
#screeplot(pca.my_cars)
library(ggplot2)
biplot(pca.my_cars)

head(pca.my_cars$scores)
max(pca.my_cars$scores[, 1])

qplot(x=my_cars$mpg, y=pca.my_cars$scores[, 1]) + labs(x="mpg", y="Comp. 1", title="Comp.1 vs mpg")
qplot(x=my_cars$mpg, y=pca.my_cars$scores[, 2]) + labs(x="mpg", y="Comp. 2", title="Comp.2 vs mpg")

cor(my_cars$mpg, pca.my_cars$scores[, 1])
```

```r
#PCA by prcomp()
pca.my_cars2 <- prcomp(my_cars[, 1:7], retx=TRUE, tol=0.5)
pca.my_cars2
summary(pca.my_cars2)
pca.my_cars2$rotation #loadings
plot(pca.my_cars2)
```

```r
#FACTOR ANALISYS
#One Factor
attach(my_cars)
my_cars_f <- cbind(mpg, cyl, disp, hp, drat, wt, qsec, vs, am, gear, carb)
fc.my_cars <- factanal(my_cars_f, factors=1, start=NULL, scores="Bartlett", STATISTIC)
fc.my_cars

head(fc.my_cars$scores, n=32)
hist(fc.my_cars$scores)
```

```r
#Two Factors with varimax-rotation
fc.my_cars <- factanal(my_cars_f, factors=2, start=NULL, scores="Bartlett", rotation="varimax", STATISTIC)
fc.my_cars
head(fc.my_cars$scores)
plot(fc.my_cars$scores)
```

```r
#Two Factors with promax-rotation
fc.my_cars <- factanal(my_cars_f, factors=2, start=NULL, scores="Bartlett", rotation="promax", STATISTIC)
fc.my_cars
head(fc.my_cars$scores)
plot(fc.my_cars$scores)
```
