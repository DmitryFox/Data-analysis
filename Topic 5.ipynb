```r
#*******Canonical ANALISYS (PCA)****************
jr<-data.frame(USJudgeRatings)
summary(jr)
cor(jr)
```

```r
# Set X (depended variables)
x<-as.matrix(USJudgeRatings[5:10])
x<-cbind(as.matrix(USJudgeRatings[1]),x)
x
```

```r
# Set Y (independed variables)
y<-as.matrix(USJudgeRatings[2:4])
y<-cbind(y,as.matrix(USJudgeRatings[11]))
y
```

```r
# Set X Statistics 
summary(x)
cx<-cor(x)
cx
```

```r
# Set Y Statistics 
summary(y)
cy<-cor(y)
cy
```

```r
# Canonicfl Correlations
cxy<-cancor(x, y)
cxy
summary(cxy)
```

```r
# Statistic Significance
library("CCP")
N = dim(x)[1] # number of observations
p = dim(x)[2] # number of dependent variables
q = dim(y)[2] # number of independent variable
N;p;q
```
